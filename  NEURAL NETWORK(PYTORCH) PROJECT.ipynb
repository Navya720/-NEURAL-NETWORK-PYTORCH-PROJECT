```python
#The  required library are listed below:

import torch as th
import torch.optim as op
import torch.nn as nn
```

```python
# numpy base library
import numpy as np
```

```python
# pands base library
import pandas as pd
```

```python
# extracting the data 

dt=np.loadtxt("pima-indians-diabetes.data.csv",delimiter=",")
dt

# load the dataset, split into input (X) and output (y) variables
```

```python
x=dt[:,0:8] #  “:“. You can select the first eight columns from index 0 to index 7 via the slice 0:8.

y=dt[:,8] # You can then select the output column (the 9th variable) via index 8.
```

```python
x,y
```

```python
# shape of the x before tranforming to tensor
x.shape
```

```python
#To convert, create a tensor out of NumPy arrays:

X=th.FloatTensor(x)
Y=th.FloatTensor(y.reshape(-1,1))
```

```python
X,Y
```

```python
X.shape # shape of the x after tranfroming into to tensor
```

```python
# define the model
```

```python
mdl=nn.Sequential(
    nn.Linear(8,10),
    nn.ReLU(),
    nn.Linear(10,6),
    nn.ReLU(),
    nn.Linear(6,1),
    nn.Sigmoid()
)
```

```python
mdl
```

```python
class LinearRgression(nn.Module):
    def __init__(self):          #  __init__ sets up the network's structure by defining the layers
        super(). __init__()
        
        self.hidden1 = nn.Linear(8, 12)
        self.act1 = nn.ReLU()
        self.hidden2 = nn.Linear(12, 8)
        self.act2 = nn.ReLU()
        self.output = nn.Linear(8, 1)
        self.act_output = nn.Sigmoid()
        
        def forward(self,x):   #  while forward specifies how the data flows through the network
        x = self.act1(self.hidden1(x))
        x = self.act2(self.hidden2(x))
        x = self.act_output(self.output(x))
        return x

model=LinearRgression()
print(model)
```

```python
# Preparation for Training

#  Adam is used. This popular version of gradient descent can automatically tune 
#  itself and gives good results in a wide range of problems.
```

```python
crietrion=nn.BCELoss() #  beacuse it's binary classification problem so we used BCELoss
optimizer=op.Adam(mdl.parameters(),lr=0.01) 
```

```python
# The goal of training a model is to ensure it learns a good enough mapping of input data to output classification.
```

```python
#The simplest way to build a training loop is to use two nested for-loops, one for epochs and one for batches:

n_epochs=10 # Epoch: Passes the entire training dataset to the model once
batch_size=5 # Batch: breaking the largedata into batchs

for epoch in range(n_epochs):
    for k in range(0,len(X),(batch_size)):
        
        Xbatch=X[k:k+batch_size]
        Ybatch=Y[k:k+batch_size]
        
        YP=mdl(Xbatch)
        loss=crietrion(YP,Ybatch)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        print(f"finshed epoch {epoch},latest loss {loss}")
```

```python
# Evaluate the Model
    
y_pred=mdl(X)
Accuracy=(y_pred.round() == Y).float().mean()
print(f" accuracy {accuracy}")
```

```python
# make probability predictions with the model
predictions = mdl(X)
# round predictions
rounded = predictions.round()
```
